```python
import matplotlib.pyplot as plt
def cool_func(x):
    return x**2+5


import numpy as np
x = np.linspace(-1, 1, 100)
plt.plot(x, cool_func(x))
plt.ylabel('y')
plt.xlabel('x')
plt.title('x^2+5')
plt.show()
x = np.linspace(-6, 6, 100)
plt.ylabel('y')
plt.xlabel('x')
plt.title('x^2+5')
plt.plot(x, cool_func(x))
plt.show()
x = np.linspace(0, 5, 100)
plt.ylabel('y')
plt.xlabel('x')
plt.title('x^2+5')
plt.plot(x, cool_func(x))
plt.show()
```
